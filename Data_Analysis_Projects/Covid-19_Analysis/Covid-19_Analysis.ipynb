## Step 1:
Importing necessary libraries.

```python
# Data analysis and Manipulation
import pandas as pd
import numpy as np

# Data Visualization
import matplotlib.pyplot as plt

# Importing Plotly
import plotly.offline as py
py.init_notebook_mode(connected=True)

# Initializing Plotly
pio.renderers.default = 'colab'
```

## Step 2:
Import datasets.

```python
# Import Dataset01
dataset01 = pd.read_csv('Downloads/covid.csv')

# Information about Dataset01
dataset01.info()
```

```python
# Import Dataset02
dataset02 = pd.read_csv('Downloads/covid_grouped.csv')

# Information about Dataset02
dataset02.info()
```

## Step 3:
Dataset cleaning.

```python
# Drop NewCases, NewDeaths, NewRecovered rows from Dataset01

dataset01.drop(['NewCases', 'NewDeaths', 'NewRecovered'], 
              axis=1, inplace=True)

# Select random set of values from Dataset01
dataset01.sample(5)
```

## Step 4:
Bar graphs- Comparisons between COVID infected countries in terms of total cases, total deaths, total recovered & total tests

```python
top15 = dataset01.head(15)
# Create graph
plt.figure(figsize=(10, 6))

# Create bargraph with 'Country/Region' column on x-axis and 'TotalCases' column on y-axis
plt.bar(top15['Country/Region'], top15['TotalCases'], color=plt.cm.viridis(top15['TotalCases'] / top15['TotalCases'].max()))

# Legends of graph
plt.title('Total Cases by Country/Region')
plt.xlabel('Country/Region')
plt.ylabel('TotalCases')

# Appearance of graph
plt.xticks(rotation=45, ha='right')  
plt.tight_layout() 

# Show graph
plt.show()
```

```python

# Create color based on 'TotalDeaths' column
colors = plt.cm.viridis(np.linspace(0, 1, len(top15)))

# Create graph
plt.figure(figsize=(10, 6))

# Create bar graph
bars = plt.bar(top15['Country/Region'], top15['TotalCases'], color=colors)

# Legends
plt.title('Total Cases by Country/Region', fontsize=14)
plt.xlabel('Country/Region', fontsize=12)
plt.ylabel('Total Cases', fontsize=12)

# Show numbers above every bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50000, f'{int(yval):,}', ha='center', va='bottom', fontsize=10)


plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show bar graph
plt.show()
```

```python
# Create graph
plt.figure(figsize=(10, 6))

# Create horizontal bar graph
bars = plt.barh(top15['Country/Region'], top15['TotalTests'], color=colors)

# Legends
plt.title('Total Tests by Country/Region', fontsize=14)
plt.xlabel('Total Tests', fontsize=12)
plt.ylabel('Country/Region', fontsize=12)

# Show number on the side of every bar
for bar in bars:
    plt.text(bar.get_width() + 50000, bar.get_y() + bar.get_height()/2, f'{int(bar.get_width()):,}', 
             va='center', fontsize=10)


plt.tight_layout()

# Show bar graph
plt.show()
```

## Step 5:
Data Visualization with scatter plot-Continent Wise

```python
# Choose the first 20 lines
top20 = dataset01.head(20)

# Define size
sizes = top20['TotalCases'] / top20['TotalCases'].max() * 80  # Normalization

# Define colors based on 'TotalCases' column
colors = top20['TotalCases']

# Create scatter plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(top20['Continent'], top20['TotalCases'], c=colors, s=sizes, cmap='viridis', alpha=0.7)

# Legends
plt.title('Total Cases by Country/Region', fontsize=14)
plt.xlabel('Country/Region', fontsize=12)
plt.ylabel('Total Cases', fontsize=12)

# Colorbar
plt.colorbar(scatter, label='Total Cases')


plt.xticks(rotation=90, ha='right')
plt.tight_layout()

# Show graph
plt.show()
```

## Step 7:
Advanced Data Visualization- Bar graphs for All top infected Countries

```python
dataset02['Date'] = pd.to_datetime(dataset02['Date'])

# Create graph
plt.figure(figsize=(10, 6))

# Color of bars based on 'Confirmed' column
bars = plt.bar(dataset02['Date'], dataset02['Confirmed'], color=plt.cm.viridis(dataset02['Confirmed'] / dataset02['Confirmed'].max()))

# Legends
plt.title('Confirmed Cases Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Confirmed Cases', fontsize=12)

# Chromatic scale 
sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=dataset02['Confirmed'].min(), vmax=dataset02['Confirmed'].max()))
sm.set_array([]) 
plt.colorbar(sm, label='Confirmed Cases')


plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show graph
plt.show()
```
