## Import libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

## Import dataset

```python
lj_playoffs = pd.read_csv("Downloads/Lebron_James_Playoffs_Career.csv")
lj_playoffs.head(5)
```

## Info about dataset


```python
lj_playoffs.info()
```

## Remove seasons that Lebron James didn't make it to the playoffs

```python
lj = pd.DataFrame(lj_playoffs)
lj_cleaned = lj.dropna(how = 'all')
lj_cleaned
```

## Total points each playoffs season using matplotlib

```python
points = np.array(lj_cleaned['PTS'])

plt.plot(points, c = 'hotpink')
```

## Total points each playoffs season using seaborn

```python
sb.lineplot(data=lj_cleaned, x="Season",y="PTS")
```

## Total rebounds each playoffs season

```python
rebounds = np.array(lj_cleaned['TRB'])

plt.plot(rebounds, c = 'magenta')
```

## Triple doubles based on the position Lebron James played

```python
sb.barplot(data=lj_cleaned, x="Pos", y="Trp-Dbl")
```

## 2-Point Field Goal Percentage throughout Lebron's playoffs career 

```python
TWO_FG_PER = np.array(lj_cleaned['2P%'])

plt.plot(TWO_FG_PER, c = 'red')
```

## 3-Point Field Goal Percentage throughout Lebron's playoffs career 

```python
THREE_FG_PER = np.array(lj_cleaned['3P%'])

plt.plot(THREE_FG_PER, c = 'lightblue')
```

## Free Throw Percentage throughout Lebron's playoffs career

```python
FT_PER = np.array(lj_cleaned['FT%'])

plt.plot(FT_PER, c='orange')
```
